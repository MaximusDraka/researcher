```python
#https://tecoholic.github.io/ner-annotator/

import spacy
import sys  
sys.path.insert(1, 'C:/Users/tom/projects/skill-skeleton/utils/')

import pandas as pd
import json

def process_data(nlp, input_file, output_file):
       
    data = {"classes":[],"annotations":[]}  
    df = pd.read_csv(input_file,delimiter='|',dtype={'data': str})
    
      
    for index,row in df.iterrows():   
                    
        text = row['data']
        
        print(index,text)     
        
        doc = nlp(text)      
        
        entities = {"entities":[]}            
        
        
        for ent in doc.ents:                
                            
            entities["entities"].append([ent.start_char,ent.end_char,ent.label_])                        
        
            if ent.label_ not in data["classes"]:
                data["classes"].append(ent.label_)
        
        data["annotations"].append([text,entities])

    with open(output_file, 'w') as f:
        json.dump(data, f, indent=4) 
   
```

```python
nlp = spacy.load('C:/Users/tom/projects/skill-skeleton/models/NER/model-best-C_PT_E_HP_P')

process_data(nlp, "vaia-data-NER.csv", "vaia-data-NER.json")
```
