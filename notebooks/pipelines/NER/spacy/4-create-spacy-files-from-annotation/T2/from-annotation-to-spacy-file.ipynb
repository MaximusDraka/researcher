```python
import spacy
import json
import sys
import os
from dotenv import load_dotenv
project_dir = os.getenv("PROJECT_PATH")

sys.path.insert(1, project_dir + '/utils/')

import spacy_util
from sklearn.model_selection import train_test_split


input_files = ["vaia-annotated-IT2.json", 
               "indeed-annotated-DA-IT2.json",
               "indeed-annotated-DE-IT2.json",
               "indeed-annotated-DS-IT2.json",
               "livecareer-annotated-DA-IT2.json",
               "livecareer-annotated-DE-IT2.json",
               "livecareer-annotated-DS-IT2.json"]

joined_train_file = "IT2-train-data.json"
joined_test_file = "IT2-test-data.json"
joined_crossvalidation_file = "IT2-crossvalidation-data.json"

output_file_train = "IT2-train-data.spacy"
output_file_test = "IT2-test-data.spacy"
output_file_valid = "IT2-valid-data.spacy"

train_pct = 0.75
test_pct = 0.15
cv_pct = 0.10

nlp = spacy.load("en_core_web_sm")

#Split each input source file into train, test and cross validation

joined_train_data = []
joined_test_data = []
joined_cv_data = []

for source_file in input_files:
    with open(source_file, 'r') as f:
        data = json.load(f)
    
    print(source_file + ' : ' + str(len(data['annotations'])))
        
    # Split the data into training and remaining data
    train_data, remaining = train_test_split(data['annotations'], test_size=1-train_pct, random_state=42)

    # Split the remaining data into testing and cross-validation data
    test_data, cv_data = train_test_split(remaining, test_size=cv_pct/(test_pct + cv_pct), random_state=42)

    # Append the data to the joined data lists
    joined_train_data.extend(train_data)
    joined_test_data.extend(test_data)
    joined_cv_data.extend(cv_data)

    print(f'Training data: {len(train_data)} examples')
    print(f'Testing data: {len(test_data)} examples')
    print(f'Cross-validation data: {len(cv_data)} examples')
    print('---------------------------')

print(f'Total Training data: {len(joined_train_data)} examples')
print(f'Total Testing data: {len(joined_test_data)} examples')
print(f'Total Cross-validation data: {len(joined_cv_data)} examples')

with open(joined_train_file, 'w') as f:
    json.dump(joined_train_data, f, indent=4)

with open(joined_test_file, 'w') as f:
    json.dump(joined_test_data, f, indent=4)

with open(joined_crossvalidation_file, 'w') as f:
    json.dump(joined_cv_data, f, indent=4)
```

```python
#Manually remove icons from the data


#Create the train data spacy file
db = spacy_util.from_ner_annotator_to_spacy(nlp,joined_train_file)
#spacy_util.print_spacy(nlp,db)
db.to_disk(output_file_train)

#Create the test data spacy file
db = spacy_util.from_ner_annotator_to_spacy(nlp,joined_test_file)
#spacy_util.print_spacy(nlp,db)
db.to_disk(output_file_test)

#Create the test data spacy file
db = spacy_util.from_ner_annotator_to_spacy(nlp,joined_crossvalidation_file)
#spacy_util.print_spacy(nlp,db)
db.to_disk(output_file_valid)
```
