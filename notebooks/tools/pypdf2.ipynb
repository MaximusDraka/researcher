```python
from spacy import displacy
import spacy
from collections import Counter

import sys
import os
from dotenv import load_dotenv

project_dir = os.getenv("PROJECT_PATH")

sys.path.insert(1, project_dir + '/utils')

import pdf_util as pdf_util


file_content = pdf_util.read_PDF(project_dir + '/data/cv/Consultant-1.pdf')

nlp=spacy.load('en_core_web_sm')
doc = nlp(file_content)
list_ent = [ent.label_ for ent in doc.ents]
labels, values = zip(*Counter(list_ent).items())

pdf_util.show_PDF_as_bar_chart(labels, values)


displacy.render(nlp(doc.text),style='ent', jupyter=True)
```
