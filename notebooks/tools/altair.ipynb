```python
from altair import *
import pandas as pd
 
data = pd.DataFrame({'a': ['Data Analyst', 'Data Engineer', 'Data Scientist'],
                     'b': [65, 42, 12]})
    

Chart(data).mark_bar().encode(
    x='a',
    y='b'
).show()

'''
df = pd.DataFrame([['Data Analyst', 5, 'Count'], 
                   ['Data Scientist', 3, 'Count'], 
                   ['Data Engineer', 4, 'Count'], 
                   ['Data Analyst', 96, 'Max'], 
                   ['Data Scientist', 99, 'Max'], 
                   ['Data Engineer', 148, 'Max'],
                     ['Data Analyst', 5.2, 'Coverage'], 
                     ['Data Scientist', 3, 'Coverage'], 
                     ['Data Engineer', 2.7, 'Coverage'],
                     ['Data Analyst', 116, 'Total score'], 
                     ['Data Scientist', 36, 'Total score'], 
                     ['Data Engineer', 62, 'Total score'] 
                  ], 
                  columns=['Profiles', 'Score', 'Genre'])
'''
```

```python
df1 = pd.DataFrame({'Profiles': {0: 'Data Analyst', 1: 'Data Scientist', 2: 'Data Engineer'},
                   'Count': {0: 5, 1: 3, 2: 4},
                   'Max': {0: 96, 1: 99, 2: 148},
                   'Coverage': {0: 5.2, 1: 3, 2: 2.7},
                   'Total score': {0: 116, 1: 36, 2: 62}}
                   )

df1
```

```python
df = pd.melt(df1, id_vars=['Profiles'], value_vars=['Count', 'Max', 'Coverage', 'Total score'], var_name='Genre', value_name='Score')
df
```

```python
chart = Chart(df).mark_bar().encode(
   column=Column('Genre'),
   x=X('Profiles'),
   y=Y('Score'),
   color=Color('Profiles', scale=Scale(range=['#EA98D2', '#659CCA', '#F6C85F']))
).configure_view(
    strokeWidth=0.0,
)

chart.show()
```
