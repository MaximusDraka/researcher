```python
import matplotlib.pyplot as plt
import networkx as nx


#############################################################################
#
# This is a Skill - Profile - Job- CV - Course ontology graph build with networkx    
#
#############################################################################


G = nx.DiGraph()
plt.figure(3,figsize=(18,10))

G.add_edge('Course', 'Skill', type='OFFERS')
G.add_edge('CV', 'Skill', type='CONTAINS')
G.add_edge('CV', 'Profile', type='IS')
G.add_edge('Profile', 'Skill', type='HAS-(Score)')
G.add_edge('Job', 'Skill', type='CONTAINS')
G.add_edge('Job', 'Profile', type='IS')
G.add_edge('Course', '1', type='annotated_by')
G.add_edge('Course', '2', type='course_info')
G.add_edge('Course', '3', type='constraints')
G.add_edge('Course', '4', type='data')
G.add_edge('Course', '5', type='date')
G.add_edge('Course', '6', type='details')
G.add_edge('Course', '7', type='full_body')
G.add_edge('Course', '8', type='intro')
G.add_edge('Course', '9', type='language')
G.add_edge('Course', '10', type='location_detail')
G.add_edge('Course', '11', type='price')
G.add_edge('Course', '12', type='start_time')
G.add_edge('Course', '13', type='sub_title')
G.add_edge('Course', '14', type='subscription_limit')
G.add_edge('Course', '15', type='target_group')
G.add_edge('Course', '16', type='title')
G.add_edge('Course', '17', type='url')
G.add_edge('Skill', '18', type='category')
G.add_edge('Skill', '19', type='name')
G.add_edge('Job', '20', type='content')
G.add_edge('Job', '21', type='url')
G.add_edge('Profile', '22', type='name')
G.add_edge('CV', '23', type='id')
G.add_edge('CV', '24', type='resume')
G.add_edge('CV', '25', type='url')


pos = nx.nx_agraph.graphviz_layout(G)
ax = plt.gca()
ax.margins(0.02)

nx.set_node_attributes(G, {"Skill": "green", "Profile": "blue", "Job": "red", "CV": "purple", "Course": "orange", "1": "grey", "2": "grey", "3": "grey", "4": "grey", "5": "grey", "6": "grey", "7": "grey", "8": "grey", "9": "grey", "10": "grey", "11": "grey", "12": "grey", "13": "grey", "14": "grey", "15": "grey", "16": "grey", "17": "grey", "18": "grey", "19": "grey", "20": "grey", "21": "grey", "22": "grey", "23": "grey", "24": "grey", "25": "grey"}, name="color")
nx.set_node_attributes(G, {"Skill": 1000, "Profile": 1500, "Job": 400, "CV": 400, "Course": 1000, "1": 200, "2": 200, "3": 200, "4": 200, "5": 200, "6": 200, "7": 200, "8": 200, "9": 200, "10": 200, "11": 200, "12": 200, "13": 200, "14": 200, "15": 200, "16": 200, "17": 200, "18": 200, "19": 200, "20": 200, "21": 200, "22": 200, "23": 200, "24": 200, "25": 200}, name="size")

# nodes
nx.draw_networkx_nodes(G, pos, node_size=[node[1] for node in G.nodes(data="size")], alpha=1, node_color=[node[1] for node in G.nodes(data="color")])

# node labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family="Verdana", font_color="white")

# edges
curved_edges = [edge for edge in G.edges() if reversed(edge) in G.edges()]
straight_edges = list(set(G.edges()) - set(curved_edges))
nx.draw_networkx_edges(G, pos, ax=ax, edgelist=straight_edges, edge_color="black", style="dashed")
arc_rad = 0.25
nx.draw_networkx_edges(G, pos, ax=ax, edgelist=curved_edges, connectionstyle=f'arc3, rad = {arc_rad}', edge_color="black", style="dashed")

# edge type labels
edge_labels = nx.get_edge_attributes(G, "type")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_family="Verdana")

plt.axis("off")
plt.show()
```
